```python
#q8 In loc function we have to give column name  whereas in iloc function we have to give index value only
```

```python
#example
import pandas as pd
data = {'A': ['a','b','c'], 'B': [40, 50, 60]}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'])
value_loc = df.loc['row2', 'B']
print(" loc:", value_loc)  

value_iloc = df.iloc[0, 1]
print("iloc:", value_iloc)  
```

```python
#Q10#groupby() is used to group the data category wise and used to perform sum,mean and etc function on group 
```

```python
#example
a= {'class': ['a', 'b', 'c'],
        'age': [26, 30, 12]}
d = pd.DataFrame(a)
group = d.groupby('class')['age'].sum()
group
```

```python
#q11-value_count is generally used to count frequency of data in a series or dataframe
data = {'C': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C']}
df = pd.DataFrame(data)

value = df['C'].value_counts()
value
```

```python
#count_plot is generally analyse data categoriacal way
import seaborn as sns
import matplotlib.pyplot as plt

data = {'C': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C']}
df = pd.DataFrame(data)
sns.countplot(x='C', data=df) 
plt.show()
```
