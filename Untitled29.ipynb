# AuearningtoTS means Automatic Time Series ,it  is a machine l

```python
df=df[["Date","Close"]]
df["Date"]=pd.to_datetime(df.Date)
df["Close"].plot(figsize=(12,8),title="Apple Stock Prices",fontsize=20,label="Close Price")
plt.legend()
plt.grid()
plt.show()
```

```python
from autots import AutoTS
model=AutoTS(forecast_length=10,frequency='infer',
            ensemble='simple',drop_data_older_than_periods=200)
model=model.fit(df,date_col='Date',value_col='Close',id_col=None)
```

```python
prediction=model.predict()
forecast=prediction.forecast
print("Stock Price Prediction of Apple")
print(forecast)
```
