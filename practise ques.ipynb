```python
#q6

yearly_temperatures = [
    55, 57, 60, 62, 65, 70, 75, 80, 78, 70, 65, 58,  # January
    60, 62, 65, 70, 75, 82, 85, 87, 80, 75, 68, 63,# February
     60, 62, 65, 70, 75, 82, 85, 87, 80, 75, 68, 63#march
    # Add daily temperatures for the remaining months
]


monthly_temperatures = {
    "January": [],
    "February": [],
    "March": [],
    
}

# Populate the monthly temperatures dictionary
month_index = 0  # Index to keep track of the current month
for temperature in yearly_temperatures:
    # Get the month name based on the current index
    month_name = list(monthly_temperatures.keys())[month_index]
    
    # Add the daily temperature to the corresponding month
    monthly_temperatures[month_name].append(temperature)
    
    # Update the month index, cycling back to 0 for January after December
    month_index = (month_index + 1) % 12

# Calculate the average temperature for each month and store it in a new dictionary
average_monthly_temperatures = {}
for month, temperatures in monthly_temperatures.items():
    average_temperature = sum(temperatures) / len(temperatures)
    average_monthly_temperatures[month] = average_temperature

# Print the average monthly temperatures
for month, average_temp in average_monthly_temperatures.items():
    print(f"{month}: {average_temp:.2f} Â°F")


    

```

```python
#q7

employee_data = [
    {"name": "Alice", "salary": 50000, "department": "HR"},
    {"name": "Bob", "salary": 60000, "department": "Finance"},
    {"name": "Charlie", "salary": 55000, "department": "HR"},
    {"name": "David", "salary": 70000, "department": "IT"},
    {"name": "Eve", "salary": 75000, "department": "IT"},
    {"name": "Frank", "salary": 62000, "department": "Finance"},
   
]

employee_by_department = {}

for employee in employee_data:
    department = employee["department"]
    if department not in employee_by_department:
        employee_by_department[department] = []  
    employee_by_department[department].append(employee["name"])  


for department, employees in employee_by_department.items():
    print(f"Department: {department}")
    print("Employees:", ", ".join(employees))
    print()

    
```

```python
#Q8

student_attendance_data = [
    {"name": "A", "attendance": [True, True, True, True, True]}, 
    {"name": "B", "attendance": [True, False, True, True, False]},  
    {"name": "C", "attendance": [True, True, True, True, False]},  #
    
]

student_attendance_percentage = {}


for student in student_attendance_data:
    name = student["name"]
    attendance = student["attendance"]
    total_days = len(attendance)
    present_days = attendance.count(True)
    
    
    attendance_percentage = (present_days / total_days) * 100
    student_attendance_percentage[name] = attendance_percentage

for name, percentage in student_attendance_percentage.items():
    print(f"{name}: {percentage:.2f}%")
```

```python
#Library Checkout System You want to create a library checkout system. Use a dictionary to store available books and their quantities. Implement a loop to allow users to check out and return books.

lib_books={'Book1':5,
          'Book2':6,
          'Book3':7,
          'Book4':3}

while True:
    print("\n Library Catlog")
    for book,quantity in lib_books.items():
        print(f"{book}-Available{quantity}")
        
        
        print("\n Options")
        print("1.check_out books")
        print("2.return_book")
        print("3.Exit")
        
        choice=int(input("Enter your choice"))
        if choice==1:
            book_checkout=input("Enter book name")
            if book_checkout in lib_books:
                if lib_books[book_checkout]>0:
                    lib_books[book_checkout]-=1
                    print("you checkout the book")
                else:
                    print("book is out of stock right now")
            else:
                print("book not found")
                
        elif choice==2:
            book_return=input("enter your bok name")
            if book_return in lib_books:
                lib_books[book_return]+=1
                print("book return sucessfully")
            else:
                print("book not found")
        elif choice==3:
            print("exiting function")
             break
        else:
            print(" invalid choice")
```

```python
#q11
#Product Inventory You run an online store and have a list of products with prices and quantities. Implement a loop that allows customers to add products to their shopping carts and calculates the total cost.
product_inventory = {
    "Product1": {"price": 10, "quantity": 50},
    "Product2": {"price": 19, "quantity": 30},
    "Product3": {"price": 5, "quantity": 75},
    "Product4": {"price": 15, "quantity": 20},
    
}

shopping_cart={}

total_cost = 0

while True:
    print("\nProduct Inventory:")
    for product, details in product_inventory.items():
        print(f"{product} - Price: {details['price']}, Quantity: {details['quantity']}")
    
    print("\nShopping Cart:")
    for product, quantity in shopping_cart.items():
        print(f"{product} - Quantity: {quantity}")
    
    print(f"\nTotal Cost: ${total_cost:.2f}")
    
    print("\nOptions:")
    print("1. Add a product to the shopping cart")
    print("2. Exit")
    
    choice = input("Enter your choice (1/2): ")
    
    if choice == "1":
        product_name = input("Enter the name of the product you want to add: ")
        if product_name in product_inventory:
            quantity_to_add = int(input("Enter the quantity you want to add: "))
            if quantity_to_add <= product_inventory[product_name]["quantity"]:
                if product_name in shopping_cart:
                    shopping_cart[product_name] += quantity_to_add
                else:
                    shopping_cart[product_name] = quantity_to_add
                product_inventory[product_name]["quantity"] -= quantity_to_add
                total_cost += product_inventory[product_name]["price"] * quantity_to_add
                print(f"{quantity_to_add} {product_name}(s) added to the shopping cart.")
            else:
                print("Sorry, the requested quantity is not available.")
        else:
            print("Product not found in the inventory.")
    
    elif choice == "2":
        print("Exiting the shopping system. Thank you for shopping with us.")
        break
    
    else:
        print("Invalid choice. Please select a valid option (1/2).")
```
