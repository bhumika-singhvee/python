```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("C:\\Users\\saurabh\\Desktop\\pyth\\covid_toy.csv")
```

```python
df.head() 
```

```python
df = df.dropna() 
```

```python
df.isnull().sum() 
```

```python
from sklearn.preprocessing import LabelEncoder 

lb = LabelEncoder() 
df['gender'] = lb.fit_transform(df['gender'])
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
x = df.drop(columns = ['has_covid'] , axis = 1)  #Independent data 
y = df['has_covid']  #Dependent data 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
from sklearn.preprocessing import FunctionTransformer 
```

```python
trf = FunctionTransformer(func = np.log1p)

x_train_transformed = trf.fit_transform(x_train) 
x_test_transformed = trf.transform(x_test) 
```

```python
import requests 

from bs4 import BeautifulSoup 
```

```python
row_data = requests.get("https://www.flipkart.com/search?q=5g%20mobiles%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
soup = BeautifulSoup(row_data.content) 
```

```python
# soup
```

```python
# Data ==> Price  , ratings , model 
```

# step - 1 (Extract Name class and create DataFrame) 

```python
x = soup.select("._4rR01T") 
name = [] 
for i in x : 
    name.append((i.contents[0]))
```

```python
# x
```

```python
# name
```

```python
df_Name = pd.DataFrame(name , columns = ['Details']) 
```

```python
df_Name
```

# Step -2 ( Extract Price Class And Create DataFrame) 

```python
x = soup.select("._30jeq3")
price = [] 
for i in x : 
    price.append((i.contents[0]))
```

```python
# x
```

```python
# price
```

```python
df_price = pd.DataFrame(price , columns = ['price'])
```

```python
df_price
```

# step-3 ( Extract Review Class And Create DataFrame) 

```python
x = soup.select("._3LWZlK") 
review = [] 
for i in x : 
    review.append((i.contents[0]))
```

```python
# x
```

```python
# review
```

```python
df_ratings = pd.DataFrame(review , columns = ['ratings'])
```

```python
df_ratings
```

# step-4 ( Concat All DataFrame) 

```python
df = pd.concat((df_Name , df_price , df_ratings) , axis = 1)
```

```python
df
```

```python
df  =  df.dropna() 
```
