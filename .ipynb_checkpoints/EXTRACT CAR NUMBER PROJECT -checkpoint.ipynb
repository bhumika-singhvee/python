```python
from PIL import Image
import pytesseract

class OCR:
    def __init__(self, tesseract_cmd):
        # Set the path to the Tesseract executable
        pytesseract.pytesseract.tesseract_cmd = tesseract_cmd

    def extract_text_from_image(self, image_path):
        try:
            # Load an image using PIL (Pillow)
            image = Image.open(image_path)

            # Perform OCR on the image
            text = pytesseract.image_to_string(image)

            return text
        except Exception as e:
            return str(e)

# Example usage:
if __name__ == "__main__":
    # Replace with the actual path to your Tesseract executable
    tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

    # Create an OCR object
    ocr = OCR(tesseract_cmd)

    # Replace 'sample_image.png' with the path to your image
    image_path = 'car.jpg'
    
    # Extract text from the image
    extracted_text = ocr.extract_text_from_image(image_path)
    
    # Print the extracted text
    print("car number is:",extracted_text)
```

```python
import pytesseract
import cv2
import re

# Step 1: Install Tesseract and required Python libraries
# You can install pytesseract using pip:
# pip install pytesseract
# Make sure you have Tesseract OCR installed on your system as well.

# Step 2: Load the image and preprocess it (if needed)
def preprocess_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)
    
    # If preprocessing is needed, you can apply various techniques like resizing, thresholding, etc.
    # For example, you can resize the image to a smaller size for faster OCR processing:
    image = cv2.resize(image, None, fx=0.5, fy=0.5)
    
    return image

# Step 3: Use Tesseract to extract text from the image
def extract_text_from_image(image):
    # Use pytesseract to extract text from the image
    extracted_text = pytesseract.image_to_string(image)
    return extracted_text

# Step 4: Apply regular expressions to find the car number
def find_car_number(text):
    # Define a regular expression pattern to match car numbers
    car_number_pattern = r'[A-Z]{2}\d{2}\s?[A-Z]{2}\s?\d{4}'
    
    # Search for the car number pattern in the extracted text
    match = re.search(car_number_pattern, text)
    
    if match:
        return match.group()
    else:
        return "Car number not found."

if __name__ == "__main__":
    # Specify the path to the image containing the car number
    image_path = "car1.jpg"
    
    # Step 2: Preprocess the image
    preprocessed_image = preprocess_image(image_path)
    
    # Step 3: Extract text from the preprocessed image using Tesseract
    extracted_text = extract_text_from_image(preprocessed_image)
    
    # Step 4: Find the car number using regular expressions
    car_number = find_car_number(extracted_text)
    
    print("Extracted Text:", extracted_text)
    print("Car Number:", car_number)
```
