```python
#Data processing
transactions=[]
for i in range(0,7501):
    transactions.append([str(df.values[i,j])for j in range(0,20)])

#training dataset
from apyori import apriori
rules=apriori(transactions,
             min_support=0.003,
             min_confidence=0.2,
             min_lift=3,
             min_length=2)

#visualizing the results
a=list(rules)
result=[list(a[i][0])for i in range (0,len(a))]
result
    

    
```
