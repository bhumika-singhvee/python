```python
#Q3=wap check even no.
#without arq
def even():
    a=int(input("even no."))
    if (a%2==0):
        print("even no.")
    else:
        print("not an even no.")
even()        
```

```python
# with arg
def evenodd(a):
    if(a%2==0):
        print("even no.")
    else:
        print("odd no.")
a=int(input("enter even no."))
evenodd(a)
```
